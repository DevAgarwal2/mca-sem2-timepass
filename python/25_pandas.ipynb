```python
import pandas as pdimport numpy as npprint("Creating series from array:")arr = np.array([10, 20, 30, 40, 50])s1 = pd.Series(arr)print(s1)print("\nCreating series from lists:")data = [100, 200, 300, 400]s2 = pd.Series(data, index=['a', 'b', 'c', 'd'])print(s2)print("\nAccessing by position:")print("s2[0]:", s2[0])print("s2[1:3]:")print(s2[1:3])print("\nAccessing by label:")print("s2['a']:", s2['a'])print("s2[['a', 'c']]:")print(s2[['a', 'c']])print("\nCreating DataFrame from dictionaries:")dict1 = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [20, 21, 22]}dict2 = {'Score': [85, 90, 78], 'Grade': ['A', 'A+', 'B+']}df1 = pd.DataFrame(dict1)df2 = pd.DataFrame(dict2)df = pd.DataFrame({'Student': dict1['Name'], 'Age': dict1['Age'], 'Score': dict2['Score'], 'Grade': dict2['Grade']})print(df)print("\nConvert DataFrame to numpy array:")numpy_array = df.valuesprint(numpy_array)print("Type:", type(numpy_array))print("\nDataFrame operations:")print("Shape:", df.shape)print("Columns:", df.columns.tolist())print("Info:")print(df.info())print("\nDescribe:")print(df.describe())
```
