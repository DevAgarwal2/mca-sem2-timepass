```python
s = {1, 2, 3, 4, 5}print("Set:", s)s.add(6)print("After add(6):", s)s.update([7, 8, 9])print("After update([7,8,9]):", s)s.remove(3)print("After remove(3):", s)s.discard(10)print("After discard(10):", s)s.pop()print("After pop():", s)s.clear()print("After clear():", s)s1 = {1, 2, 3, 4, 5}s2 = {3, 4, 5, 6, 7}print("\nSet Operations:")print("s1:", s1)print("s2:", s2)print("Union:", s1 | s2)print("Intersection:", s1 & s2)print("Difference:", s1 - s2)print("Symmetric Difference:", s1 ^ s2)print("\nFrozenset:")fs = frozenset([1, 2, 3])print("Frozenset:", fs)
```
