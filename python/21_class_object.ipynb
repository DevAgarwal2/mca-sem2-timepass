```python
class Person:    def __init__(self, name, age):        self.name = name        self.age = age    def display(self):        print("Name: {}, Age: {}".format(self.name, self.age))    def greet(self):        print("Hello, {}!".format(self.name))p1 = Person("John", 25)p2 = Person("Alice", 22)print("Object 1:")p1.display()p1.greet()print("\nObject 2:")p2.display()p2.greet()print("\nAccessing attributes:")print("p1.name:", p1.name)print("p1.age:", p1.age)p1.age = 26print("After updating p1.age:", p1.age)
```
