```python
from tkinter import *def on_button_click():    label.config(text="Button Clicked!")def on_check():    label.config(text="Checkbox " + ("Checked" if var.get() else "Unchecked"))def get_entry():    label.config(text="Entry: " + entry.get())def show_selection():    label.config(text="Selected: " + listbox.get(listbox.curselection()))root = Tk()root.title("Tkinter Widgets Demo")root.geometry("400x500")label = Label(root, text="Tkinter Widgets", font=("Arial", 16, "bold"))label.pack(pady=10)Button(root, text="Click Me", command=on_button_click, bg="lightblue").pack(pady=5)var = IntVar()Checkbutton(root, text="Check me", variable=var, command=on_check).pack(pady=5)entry = Entry(root, width=30)entry.pack(pady=5)Button(root, text="Get Entry", command=get_entry).pack(pady=5)frame = Frame(root, bd=2, relief=SUNKEN)frame.pack(pady=10, padx=10)Label(frame, text="This is a Frame").pack()listbox = Listbox(root)listbox.insert(1, "Python")listbox.insert(2, "Java")listbox.insert(3, "C++")listbox.insert(4, "JavaScript")listbox.pack(pady=5)Button(root, text="Show Selection", command=show_selection).pack(pady=5)mb = Menubutton(root, text="Menu", relief=RAISED)mb.pack(pady=10)menu = Menu(mb, tearoff=0)mb.config(menu=menu)menu.add_command(label="Item 1")menu.add_command(label="Item 2")menu.add_separator()menu.add_command(label="Item 3")label.pack(pady=20)root.mainloop()
```
