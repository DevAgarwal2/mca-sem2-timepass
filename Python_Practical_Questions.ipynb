#### (a) if Statement - Check if number is positive

```python
# (a) if Statement - Check if number is positive
num = 5
if num > 0:
    print(num, "is a positive number")
```

#### (b) if...else Statement - Check if number is positive or negative

```python
# (b) if...else Statement - Check if number is positive or negative
num = -3
if num > 0:
    print(num, "is a positive number")
else:
    print(num, "is a negative number")
```

#### (c) if...elif...else Statement - Check if number is positive, negative or zero

```python
# (c) if...elif...else Statement
num = 0
if num > 0:
    print(num, "is a positive number")
elif num < 0:
    print(num, "is a negative number")
else:
    print(num, "is zero")
```

#### (d) Nested if - Check if number is positive, negative or zero

```python
# (d) Nested if Statement
num = 7
if num >= 0:
    if num == 0:
        print(num, "is zero")
    else:
        print(num, "is a positive number")
else:
    print(num, "is a negative number")
```

#### (e) for Loop - Sum of all numbers in a list

```python
# (e) for Loop - Sum of all numbers stored in a list
numbers = [10, 20, 30, 40, 50]
total = 0
for num in numbers:
    total += num
print("Sum of all numbers in the list:", total)
```

#### (f) while Loop - Sum of natural numbers up to n

```python
# (f) while Loop - Add natural numbers up to n
n = 10
sum_nat = 0
i = 1
while i <= n:
    sum_nat += i
    i += 1
print("Sum of natural numbers from 1 to", n, "is:", sum_nat)
```

### Question 2E: Special Number Check

```python
# Special Number Check (e.g., 59)
def is_special_number(num):
    # Convert number to string to extract digits
    digits = [int(d) for d in str(num)]
    
    # Sum of all digits
    digit_sum = sum(digits)
    
    # Multiplication of all digits
    digit_product = 1
    for d in digits:
        digit_product *= d
    
    # Addition of sum and multiplication
    result = digit_sum + digit_product
    
    return result == num

# Test with 59
num = 59
if is_special_number(num):
    print(f"{num} is a special number")
else:
    print(f"{num} is not a special number")
```

### Question 3A: Area of Triangle with Validation

```python
import math

def areaTriangle(a, b, c):
    # Assert that sum of any two sides is greater than third side
    assert a + b > c and b + c > a and a + c > b, "Invalid triangle: sum of any two sides must be greater than third side"
    
    # Calculate semi-perimeter
    s = (a + b + c) / 2
    
    # Calculate area using Heron's formula
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))
    return area

def main():
    # Get triangle sides from user
    a = float(input("Enter first side of triangle: "))
    b = float(input("Enter second side of triangle: "))
    c = float(input("Enter third side of triangle: "))
    
    try:
        result = areaTriangle(a, b, c)
        print(f"Area of triangle: {result:.2f} square units")
    except AssertionError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```

### Question 3B: Co-prime Numbers

```python
def are_co_primes(a, b):
    def gcd(x, y):
        while y:
            x, y = y, x % y
        return x
    
    return gcd(a, b) == 1

# Test the function
num1 = 14
num2 = 25
if are_co_primes(num1, num2):
    print(f"{num1} and {num2} are co-primes")
else:
    print(f"{num1} and {num2} are not co-primes")
```

### Question 3C: Successive Repetitive Characters Replaced with *

```python
def replace_repetitive(s):
    result = ""
    for i in range(len(s)):
        if i > 0 and s[i] == s[i-1]:
            result += "*"
        else:
            result += s[i]
    return result

# Test with example
test_string = "balloon"
print(f"Original: {test_string}")
print(f"Modified: {replace_repetitive(test_string)}")
```

### Question 3D: Number to Words and Kaprekar Number Check

```python
def number_to_words(n):
    digit_to_word = {
        '0': 'Zero', '1': 'One', '2': 'Two', '3': 'Three', '4': 'Four',
        '5': 'Five', '6': 'Six', '7': 'Seven', '8': 'Eight', '9': 'Nine'
    }
    return ' '.join(digit_to_word[d] for d in str(n))

def is_kaprekar_number(n):
    squared = n ** 2
    str_squared = str(squared)
    
    for i in range(1, len(str_squared)):
        left = int(str_squared[:i])
        right = str_squared[i:]
        if right:  # Make sure right part is not empty
            right = int(right)
            if left + right == n:
                return True
    return False

# Test functions
num = 452
print(f"Number to words: {number_to_words(num)}")

# Check Kaprekar numbers
print(f"\nChecking Kaprekar numbers:")
for n in [9, 297, 452, 5, 10]:
    print(f"{n} is Kaprekar: {is_kaprekar_number(n)}")
```

### Question 3E: KAPREKAR NUMBER Demonstration

```python
def is_kaprekar(n):
    squared = n ** 2
    str_squared = str(squared)
    
    for i in range(1, len(str_squared)):
        left = int(str_squared[:i]) if str_squared[:i] else 0
        right = int(str_squared[i:]) if str_squared[i:] else 0
        if left + right == n:
            return True
    return False

# Examples
print("Kaprekar Number Examples:")
print("9: 9^2 = 81, 8 + 1 = 9 => Kaprekar:", is_kaprekar(9))
print("297: 297^2 = 88209, 88 + 209 = 297 => Kaprekar:", is_kaprekar(297))
print("452: 452^2 = 204420, 204 + 420 = 624 => Kaprekar:", is_kaprekar(452))
print("45: 45^2 = 2025, 20 + 25 = 45 => Kaprekar:", is_kaprekar(45))
```

### Question 5A: Sum and Average of Digits in String

```python
import re

def sum_avg_digits(s):
    digits = [int(c) for c in s if c.isdigit()]
    if digits:
        return sum(digits), sum(digits)/len(digits)
    return 0, 0

test_string = "abc123def45"
total, average = sum_avg_digits(test_string)
print(f"String: {test_string}")
print(f"Sum of digits: {total}")
print(f"Average of digits: {average}")
```

### Question 5B: Arrange String with Lowercase First

```python
def arrange_string(s):
    lower_chars = [c for c in s if c.islower()]
    upper_chars = [c for c in s if c.isupper()]
    return ''.join(lower_chars + upper_chars)

test_string = "PyThOn"
print(f"Original: {test_string}")
print(f"Arranged: {arrange_string(test_string)}")
```

### Question 5D: Days Between Two Dates

```python
from datetime import datetime

def days_between(date1_str, date2_str):
    date1 = datetime.strptime(date1_str, "%Y-%m-%d")
    date2 = datetime.strptime(date2_str, "%Y-%m-%d")
    return abs((date2 - date1).days)

date1 = "2024-01-01"
date2 = "2024-01-15"
print(f"Days between {date1} and {date2}: {days_between(date1, date2)}")
```

### Question 5F: Subtract a Week from a Given Date

```python
from datetime import datetime, timedelta

def subtract_week(date_str):
    date = datetime.strptime(date_str, "%Y-%m-%d")
    new_date = date - timedelta(weeks=1)
    return new_date.strftime("%Y-%m-%d")

original_date = "2024-01-15"
print(f"Original date: {original_date}")
print(f"Date after subtracting 1 week: {subtract_week(original_date)}")
```

### Question 6: Copy Text File with Error Handling

```python
def copy_file():
    source_file = input("Enter source file name: ")
    dest_file = input("Enter destination file name: ")
    
    try:
        with open(source_file, 'r') as src:
            content = src.read()
        
        with open(dest_file, 'w') as dest:
            dest.write(content)
        
        print(f"File copied successfully from {source_file} to {dest_file}")
        
    except FileNotFoundError:
        print(f"Error: Source file '{source_file}' not found.")
    except PermissionError:
        print(f"Error: Permission denied to access the file.")
    except IOError as e:
        print(f"Error: An I/O error occurred: {e}")
    except Exception as e:
        print(f"Error: An unexpected error occurred: {e}")

# Test with a sample file
# copy_file()
```

### Question 7: Shape and Circle Classes

```python
import math

class Shape:
    def getPerimeter(self):
        pass
    
    def getArea(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def getPerimeter(self):
        return 2 * math.pi * self.radius
    
    def getArea(self):
        return math.pi * self.radius ** 2

# Test Circle class
circle = Circle(5)
print(f"Circle with radius {circle.radius}:")
print(f"Area: {circle.getArea():.2f}")
print(f"Perimeter: {circle.getPerimeter():.2f}")
```

### Question 8: MySQL Database Connectivity

**Note:** Install mysql-connector-python using: `pip install mysql-connector-python`

```python
# MySQL Database Connectivity
import mysql.connector
from mysql.connector import Error

def database_connectivity():
    try:
        # Connect to MySQL server (update credentials as needed)
        connection = mysql.connector.connect(
            host='localhost',
            user='root',
            password='your_password'
        )
        
        if connection.is_connected():
            print("1. Database Connection Status: CONNECTED")
            cursor = connection.cursor()
            
            # II. Create database BANK
            cursor.execute("CREATE DATABASE IF NOT EXISTS BANK")
            print("2. Database BANK created successfully")
            
            cursor.execute("USE BANK")
            
            # III. Create table
            cursor.execute("""
                CREATE TABLE IF NOT EXISTS BankBranch (
                    branch_id INT PRIMARY KEY,
                    name VARCHAR(100),
                    MobileNo VARCHAR(15),
                    Email_id VARCHAR(100),
                    IFSC VARCHAR(20),
                    City VARCHAR(50)
                )
            """)
            print("3. Table BankBranch created successfully")
            
            # IV. Insert data
            insert_query = """
                INSERT INTO BankBranch (branch_id, name, MobileNo, Email_id, IFSC, City)
                VALUES (%s, %s, %s, %s, %s, %s)
            """
            values = [(1, 'Main Branch', '9876543210', 'main@bank.com', 'BANK001', 'Mumbai')]
            cursor.executemany(insert_query, values)
            connection.commit()
            print("4. Data inserted successfully")
            
            # V. Update data
            update_query = "UPDATE BankBranch SET City = 'Delhi' WHERE branch_id = 1"
            cursor.execute(update_query)
            connection.commit()
            print("5. Data updated successfully")
            
            # VI. Delete data
            delete_query = "DELETE FROM BankBranch WHERE branch_id = 1"
            cursor.execute(delete_query)
            connection.commit()
            print("6. Data deleted successfully")
            
            # Insert fresh data for fetch operations
            cursor.executemany(insert_query, [(1, 'Main Branch', '9876543210', 'main@bank.com', 'BANK001', 'Mumbai'),
                                               (2, 'Branch 2', '9876543211', 'branch2@bank.com', 'BANK002', 'Delhi')])
            connection.commit()
            
            # VII. Fetch one record
            cursor.execute("SELECT * FROM BankBranch LIMIT 1")
            result = cursor.fetchone()
            print("7. Fetch one record:", result)
            
            # VIII. Fetch all records
            cursor.execute("SELECT * FROM BankBranch")
            results = cursor.fetchall()
            print("8. Fetch all records:")
            for row in results:
                print(row)
            
            # IX. List databases
            cursor.execute("SHOW DATABASES")
            databases = cursor.fetchall()
            print("9. Databases in MySQL:")
            for db in databases:
                print(db[0])
    
    except Error as e:
        print(f"Error: {e}")
    finally:
        if connection.is_connected():
            cursor.close()
            connection.close()
            print("\nDatabase connection closed.")

# Uncomment to run:
# database_connectivity()
```

```python
import tkinter as tk
from tkinter import messagebox

def submit_form():
    name = entry_name.get()
    roll = entry_roll.get()
    email = entry_email.get()
    phone = entry_phone.get()
    course = course_var.get()
    gender = gender_var.get()
    
    if not name or not roll or not email:
        messagebox.showerror("Error", "Please fill all required fields")
        return
    
    messagebox.showinfo("Success", f"Admission Form Submitted!\n\nName: {name}\nRoll No: {roll}\nEmail: {email}\nPhone: {phone}\nCourse: {course}\nGender: {gender}")

def clear_form():
    entry_name.delete(0, tk.END)
    entry_roll.delete(0, tk.END)
    entry_email.delete(0, tk.END)
    entry_phone.delete(0, tk.END)
    entry_address.delete("1.0", tk.END)
    course_var.set("Select Course")
    gender_var.set("")

# Create main window
root = tk.Tk()
root.title("College Admission Form")
root.geometry("500x600")

# Title Label
title = tk.Label(root, text="COLLEGE ADMISSION FORM", font=('Arial', 18, 'bold'))
title.pack(pady=20)

# Frame for form fields
form_frame = tk.Frame(root)
form_frame.pack(pady=10)

# Name Label and Entry
tk.Label(form_frame, text="Name:", font=('Arial', 10)).grid(row=0, column=0, sticky='w', pady=5)
entry_name = tk.Entry(form_frame, width=30)
entry_name.grid(row=0, column=1, pady=5)

# Roll Number
tk.Label(form_frame, text="Roll No:", font=('Arial', 10)).grid(row=1, column=0, sticky='w', pady=5)
entry_roll = tk.Entry(form_frame, width=30)
entry_roll.grid(row=1, column=1, pady=5)

# Email
tk.Label(form_frame, text="Email:", font=('Arial', 10)).grid(row=2, column=0, sticky='w', pady=5)
entry_email = tk.Entry(form_frame, width=30)
entry_email.grid(row=2, column=1, pady=5)

# Phone
tk.Label(form_frame, text="Phone:", font=('Arial', 10)).grid(row=3, column=0, sticky='w', pady=5)
entry_phone = tk.Entry(form_frame, width=30)
entry_phone.grid(row=3, column=1, pady=5)

# Course (Dropdown/OptionMenu)
tk.Label(form_frame, text="Course:", font=('Arial', 10)).grid(row=4, column=0, sticky='w', pady=5)
course_var = tk.StringVar(root)
course_var.set("Select Course")
course_menu = tk.OptionMenu(form_frame, course_var, "B.Tech", "B.Sc", "B.Com", "BBA", "MBA")
course_menu.config(width=27)
course_menu.grid(row=4, column=1, pady=5)

# Gender (Radio Buttons)
tk.Label(form_frame, text="Gender:", font=('Arial', 10)).grid(row=5, column=0, sticky='w', pady=5)
gender_var = tk.StringVar(root)
tk.Radiobutton(form_frame, text="Male", variable=gender_var, value="Male").grid(row=5, column=1, sticky='w')
tk.Radiobutton(form_frame, text="Female", variable=gender_var, value="Female").grid(row=5, column=1, sticky='e')
tk.Radiobutton(form_frame, text="Other", variable=gender_var, value="Other").grid(row=6, column=1, sticky='w')

# Address (Text Widget)
tk.Label(form_frame, text="Address:", font=('Arial', 10)).grid(row=7, column=0, sticky='nw', pady=5)
entry_address = tk.Text(form_frame, width=25, height=4)
entry_address.grid(row=7, column=1, pady=5)

# Buttons Frame
button_frame = tk.Frame(root)
button_frame.pack(pady=20)

# Submit Button
submit_btn = tk.Button(button_frame, text="Submit", command=submit_form, bg='green', fg='white', width=10)
submit_btn.grid(row=0, column=0, padx=10)

# Clear Button
clear_btn = tk.Button(button_frame, text="Clear", command=clear_form, bg='orange', fg='white', width=10)
clear_btn.grid(row=0, column=1, padx=10)

# Checkbutton for terms
terms_var = tk.BooleanVar()
terms_check = tk.Checkbutton(root, text="I agree to the terms and conditions", variable=terms_var)
terms_check.pack(pady=10)

# Run the application
root.mainloop()
```
